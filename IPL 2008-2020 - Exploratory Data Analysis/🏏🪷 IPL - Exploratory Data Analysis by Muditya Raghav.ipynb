# üèè IPL Matches Exploratory Data Analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# loading data from ipl.csv
ipl = pd.read_csv("ipl.csv")
```

```python
# size of data - row, and col
ipl.shape
ipl.head()
```

```python
# how many times, a man of the match a person awarded top 10
ipl['player_of_match'].value_counts()[:5]
```

```python
mom = list(ipl['player_of_match'].value_counts()[:5])
```

```python
plt.figure(figsize=(6,4))
plt.bar(list(ipl['player_of_match'].value_counts()[:5].keys()),list(ipl['player_of_match'].value_counts()[:5]),color="g")
```

```python
# result of matches
ipl['result'].value_counts()
```

```python
# winners
ipl['toss_winner'].value_counts()
```

```python
# Winning team which takes batting in first place
b_first = ipl[ipl['win_by_runs']!=0]
```

```python
b_first.head()
```

```python
#making histogram
plt.hist(b_first['win_by_runs'],color='g')
plt.title("Distribution by runs")
plt.xlabel("Runs")
```

```python
# winning team by batting first
b_first['winner'].value_counts()
```

```python
# bar-graph for top 3 teams -> first batting and winners
plt.figure(figsize=(6,3))
plt.bar(list(b_first['winner'].value_counts()[:3].keys()),list(b_first['winner'].value_counts()[:3]),color="g")
```

```python
# pie-charts for top 3 teams -> first batting and winners
plt.figure(figsize=(10,9))
plt.pie(list(b_first['winner'].value_counts()), labels=list(b_first['winner'].value_counts().keys()), autopct="%0.2f%%")
```

```python
# records where a team has won after second batting
b_second = ipl[ipl['win_by_wickets']!=0]
```

```python
b_second.head()
```

```python
b_second['winner'][:10]
```

```python
# pie-charts for top 3 teams -> second batting and winners
plt.figure(figsize=(10,9))
plt.pie(list(b_second['winner'].value_counts()), labels=list(b_second['winner'].value_counts().keys()), autopct="%0.2f%%")
```

```python
# frequency of wins w.r.t to number of wickets
plt.figure(figsize=(3,3))
plt.hist(b_second['win_by_wickets'], bins=30)
```

```python
# team which wins most match after 2nd batting
b_second['winner'].value_counts()
```

```python
# team which wins most match in first batting
b_first['winner'].value_counts()
```

```python
# no of matchs in season
ipl['season'].value_counts()
```

```python
# no of matchs in city
ipl['city'].value_counts()
```

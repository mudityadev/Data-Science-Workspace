```python
# OOPs concept in python

class Girl:
    def set_fig(self, bust):
        self.bust = bust
    def get_fig(self):
        return self.bust
    def set_complex(self, color):
        self.color = color
    def get_complex(self):
        return self.color
    def set_size(self, b, w, h):
        self.b = b
        self.w = w
        self.h = h
    def get_size(self):
        print("Size = ", self.b, " ", self.w, " ",self.h)
```

```python
sunny = Girl()

sunny.set_fig(38)
sunny.set_complex("dark")
sunny.get_fig()
sunny.get_complex()
sunny.set_size(38,34,36)
sunny.get_size()
```

```python
# Constructor
class Models:
    def __init__(self, color, waist):
        self.color = color
        self.waist = waist
        
    def get_details(self):
        print("color = ", self.color)
        print("waist = ", self.waist)
        
class Actoress:
    def hello(self):
        print("hey, I am actoress")
```

```python
# Inheritance
class JBPModels(Models, Actoress): # multiple inheritance
    # over-riding the constructor
    def __init__(self,color, waist, location, phone):
        super().__init__(color, bust, waist, hip) # sending to super class
        self.location = location
        self.phone = phone
    
    def jbp_details(self):
        print("location = ", self.location)
        print("phone = ", self.phone)      
        
        
nansi = JBPModels("white", 32, "jabalpur", 8888888888)
     
nansi.get_details()
nansi.jbp_details()
nansi.hello()
```

```python
class Queen(JBPModels):
    def room(self):
        print("i am a queen")
```

```python
sakshi = Queen("fair", 40,36,38,"bhopal", 989)

sakshi.get_details()
sakshi.jbp_details()
sakshi.hello()
sakshi.room()
```

```python
cost = np.array([10,20,30,40,50])
price = np.array([10,20,20,34,40])

np.intersect1d(cost, price) # common, intersected element
# array([10, 20, 40])

np.setdiff1d(cost, price) # cost - price
# array([30, 50])

np.setdiff1d(price,cost) # price - cost
# array([ 2, 21, 32, 34])

np.sum([cost, price])
# total sum

np.sum([cost, price], axis=1) # row, axis=1
# array([150, 124])

np.sum([cost, price], axis=0) # col,axis = 0
# array([20, 40, 50, 74, 90])

cost-=3
cost+=3
cost*=3
cost/=3
```

```python
np.mean(cost) # mean
# 822.0

np.std(cost) # standard deviation
# 381.8376618407357

np.median(price) # median
# 20.0
```

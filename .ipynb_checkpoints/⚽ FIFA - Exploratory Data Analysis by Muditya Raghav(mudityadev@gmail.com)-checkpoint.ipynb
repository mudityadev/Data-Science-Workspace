```python
# loading libraries
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# load the csv file
fifa = pd.read_csv("players_20.csv")
```

```python
# total players with their attributes
fifa.shape
```

```python
# data 
fifa.head()
```

```python
# total columns
for col in fifa.columns:
    print(col)
```

```python
# top 10 country which have most players
fifa['nationality'].value_counts()[:5]
```

```python
# top 10 country which have most players

fifa['nationality'].value_counts()[:5].keys()
```

```python
# bar-charts, most players from which country
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]))
```

```python
# get the salary and name of players in player_salary
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
# get the player who have highest salary 
player_salary = player_salary.sort_values(by=['wage_eur'], ascending=False)
```

```python
player_salary[:10]
```

```python
# bar-charts, the player who have highest salary 
plt.figure(figsize=(7,3))
plt.bar(list(player_salary['short_name'].value_counts()[0:5].keys()),list(player_salary['wage_eur'].value_counts()[0:5]))
```

```python
# Total players in germany
germany = fifa[(fifa['nationality'] == 'Germany')!=False]
```

```python
germany.shape
# 1216 players are from germany
```

```python
germany.sort_values(by=['height_cm'], ascending=False).head(5)
```

```python
germany.sort_values(by=['wage_eur'], ascending=False).head(5)
```

```python
germany.sort_values(by=['weight_kg'], ascending=False).head(5)
```

```python
germany[['short_name','wage_eur','weight_kg','height_cm']].sort_values(by=['wage_eur'], ascending=False).head(5)
```

```python
# Shooting Skills analysis
player_shooting = fifa[['short_name','shooting']]
```

```python
# Top shooters in football
player_shooting.sort_values(by="shooting", ascending=False)[:10]
```

```python
# defending skills
player_defend = fifa[['short_name','defending','nationality','club']]
```

```python
# Top defenders
player_defend.sort_values(by=['defending'],ascending=False).head()
```

```python
# Club-based analysis - Real Madrid
real_madrid = fifa[fifa['club'] == 'Real Madrid']
```

```python
real_madrid[['short_name','wage_eur']].sort_values(by=['wage_eur'], ascending=False).head()
```

```python
real_madrid[['short_name','shooting']].sort_values(by=['shooting'], ascending=False).head()
```

```python
real_madrid[['short_name','defending']].sort_values(by=['defending'], ascending=False).head()
```
